```python
# Create a Summary DataFrame
ts = schoolsfile["school_name"].count()  #total schools
tst = studentfile["student_name"].count()  #total students
tb = schoolsfile["budget"].sum()  #total budget
ams = studentfile["math_score"].mean()  #avg math score
ars = studentfile["reading_score"].mean()  #avg reading score
pm_df = school_data_complete[school_data_complete['math_score'] >= 70]  #pct passing math
ppm = pm_df["student_name"].count()/tst
pr_df = school_data_complete[school_data_complete['reading_score'] >= 70]  #pct passing reading
ppr = pr_df["student_name"].count()/tst
opp = (ams + ars) / 2 #overall pct passing


district_summary_df = pd.DataFrame({
    "Total Schools": [ts],
    "Total Students": [tst],
    "Total Budget": [tb],
    "Average Math Score":[ams],
    "Average Reading Score":[ars],
    "% Passing Math":[ppm],
    "% Passing Reading":[ppr],
    "Overall Passing Rate":[opp]
})
district_summary_df
```

```python
# School Summary
indx = np.arange(0,ts)
cols = ['School Name','School Type','Total Students','Total School Budget',
            'Per Student Budget','Average Math Score','Average Reading Score',
            '% Passing Math','% Passing Reading','Overall Passing Rate']
school_summary_df = pd.DataFrame(index = indx, columns = cols)
school_summary_df['School Name'] = schoolsfile['school_name']
school_summary_df['School Type'] = schoolsfile['type']
school_summary_df['Total Students'] = schoolsfile['size']
school_summary_df['Total School Budget'] = schoolsfile['budget']
school_summary_df['Per Student Budget'] = schoolsfile['budget']/schoolsfile['size']
school_summary_df['Average Math Score'] = [studentfile[studentfile.school_name == sch]['math_score'].mean() 
                                           for sch in schoolsfile['school_name']]
school_summary_df['Average Reading Score'] = [studentfile[studentfile.school_name == sch]['reading_score'].mean() 
                                              for sch in schoolsfile['school_name']]
school_summary_df['% Passing Math'] = [studentfile[(studentfile.school_name == sch) & 
                                                   (studentfile.math_score >= 70)]['math_score'].count() / 
                                       studentfile[studentfile.school_name == sch]['math_score'].count() 
                                       for sch in schoolsfile['school_name']]
school_summary_df['% Passing Reading'] = [studentfile[(studentfile.school_name == sch) & 
                                                   (studentfile.math_score >= 70)]['reading_score'].count() / 
                                       studentfile[studentfile.school_name == sch]['reading_score'].count() 
                                       for sch in schoolsfile['school_name']]
school_summary_df['Overall Passing Rate'] = (school_summary_df['% Passing Math'] + school_summary_df['% Passing Reading']) / 2
```
